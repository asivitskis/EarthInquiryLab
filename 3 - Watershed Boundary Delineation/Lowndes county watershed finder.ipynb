```python
# Import necessary libraries and set environment variable for nodata value.
import os
import leafmap
import numpy as np

os.environ["NODATA"] = "-32768"
```

```python
# Define map and central coordinates on area of interest.
m = leafmap.Map()

# Set central coordinates, for example Central HS.
lat = 32.152684183972255
lon = -86.68236184327075


m = leafmap.Map(center=[lat, lon], zoom=10)
m.add_marker([lat, lon])
m
```

```python
# Set geometry for watershed delineation.
geometry = {"x": lon, "y": lat}
```

```python
# Get watershed boundary and visualize it on the map.
gdf = leafmap.get_wbd(geometry, digit=10, return_geometry=True)
gdf.explore()
```

```python
# Save watershed boundary as GeoJSON file.
gdf.to_file("basin.geojson")
```

```python
# Get Digital Elevation Model for watershed boundary and save as Cloud Optimized GeoTIFF (COG).
array = leafmap.get_3dep_dem(
    gdf,
    resolution=30,
    output="dem.tif",
    dst_crs="EPSG:3857",
    to_cog=True,
    overwrite=True,
)
array
```

```python
# Visualize DEM on the map and add color bar.
m.add_raster("dem.tif", palette="terrain", nodata=np.nan, layer_name="DEM")
m.add_colormap(
    "terrain",
    label="Elevation",
    orientation="horizontal",
    vmin=0,
    vmax=150,
)
m
```
